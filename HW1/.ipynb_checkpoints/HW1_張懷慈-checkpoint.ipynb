```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
top_hits = pd.read_csv('Spotify 2010 - 2019 Top 100.csv')
```

```python
top_hits.head()
```

## 三年人氣王
使用聯集找出2017-2019年都有上榜的藝人，並看看這些藝人這3年上榜的次數

```python
# 分別建立 2019、2018、2017 年藝人的 set
top_artist_2019_set = set(top_hits.loc[top_hits['top year'] == 2019,'artist'])
top_artist_2018_set = set(top_hits.loc[top_hits['top year'] == 2018,'artist'])
top_artist_2017_set = set(top_hits.loc[top_hits['top year'] == 2017,'artist'])
```

```python
# 進行交集
top_artist_intersection = top_artist_2019_set & top_artist_2018_set & top_artist_2017_set
top_artist_intersection
```

```python
# 找出交集藝人2017-2019年上榜的歌曲有幾首
top_artist_3year = top_hits.loc[top_hits['artist'].isin(top_artist_intersection) & top_hits['top year'].isin([2019,2018,2017]),'artist'].value_counts().to_frame().reset_index()
top_artist_3year.columns = ['Artist', 'Count']
top_artist_3year
```

## This is trend
使用聯集找出2017-2019的上榜的音樂類型有哪些，看看三年潮流類型與最熱門音樂類型前5名

```python
# 分別建立 2019、2018、2017 年種類的 set，並進行聯集
top_type_2019_set = set(top_hits.loc[top_hits['top year'] == 2019,'top genre'])
top_type_2018_set = set(top_hits.loc[top_hits['top year'] == 2018,'top genre'])
top_type_2017_set = set(top_hits.loc[top_hits['top year'] == 2017,'top genre'])
total_type = top_type_2019_set | top_type_2018_set | top_type_2017_set
print('共有'+str(len(total_type))+'種類型音樂：')
total_type
```

```python
# 計算 2017-2019 年這些音樂種類每年在排行榜上出現的次數，並取出前 10 名
top_10_genre = top_hits.loc[top_hits['top genre'].isin(total_type) & top_hits['top year'].isin([2019,2018,2017]),'top genre'].value_counts().head(10).to_frame().reset_index()
top_10_genre.columns = ['Genre', 'Count']
top_10_genre
```

```python
# 將結果會製成圓餅圖

# 設定圖的大小
plt.figure(figsize=(12, 9))

# 設定圖的顏色
colors = ['#fbf8cc', '#fde4cf', '#ffcfd2', '#f1c0e8', '#cfbaf0', 
          '#a3c4f3', '#a3c4f3', '#8eecf5', '#98f5e1', '#b9fbc0']

# 產出圓餅圖與標題
plt.pie(top_10_genre['Count'], labels=top_10_genre['Genre'], autopct='%1.1f%%', 
        colors=colors, textprops={'fontsize': 16})
plt.title('Ranking Count by Genre', fontsize=24)

plt.tight_layout()
plt.show()
```

## 後浪推前浪
以2010-2012、2017-2019這兩個區間做比較，使用差集找出在2010's後期才出現後起新秀

```python
artist_now_set = set(top_hits.loc[top_hits['top year'].isin([2019,2018,2017]),'artist'])
artist_then_set = set(top_hits.loc[top_hits['top year'].isin([2012,2011,2010]),'artist'])
new_star = (artist_now_set - artist_then_set)
```

```python
rising_star = top_hits.loc[top_hits['artist'].isin(new_star) & top_hits['top year'].isin([2019,2018,2017]),'artist'].value_counts().head(10).to_frame().reset_index()
rising_star.columns = ['Artist', 'Count']
rising_star
```

```python
# 設定圖的大小
plt.figure(figsize=(12, 9))
# plt.bar(top_10_genre['Genre'], top_10_genre['Count'], color='#CCC9E7')

# 將前3名與其他名次定義為不同顏色
target_index = [0,1,2]
target_color = '#9c89b8'
other_color = '#b8bedd'
colors = [target_color if i in target_index else other_color for i in range(len(top_10_genre))]
plt.bar(rising_star['Artist'], rising_star['Count'], color=colors)

# 標題名稱與大小調整
plt.title('Ranking Count by Artist', fontsize=24)
plt.xlabel('Artist', fontsize=20)
plt.ylabel('Ranking Count', fontsize=20)

# 調整x軸、y軸字體大小
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

#顯示圖形
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
