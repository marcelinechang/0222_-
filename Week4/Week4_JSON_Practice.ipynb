```python
import json
import pandas as pd
```

```python
# Opening JSON file
f = open('Electric Vehicle Population Data.json')

# returns JSON object as 
# a dictionary
data = json.load(f)

# Closing file
f.close()
```

```python
data
```

```python
# 查看資料結構並找到製造商[Make]資訊是在list中的哪個位置
structure = data['meta']['view']['columns']
for _ in enumerate(structure):
    print(_[0],_[1]['name']) 
```

```python
# 計算這份資料中，每個製造商產出多少輛車子，並製作成字典
detail = data['data']
builder = [manufature[14] for manufature in detail]
builder_count = {i:builder.count(i) for i in set(builder)}
builder_count
```

```python
# 將結果轉成dataframe，並依照產出量進行排序
ranking = pd.DataFrame.from_dict(builder_count, orient='index').reset_index().rename(columns={"index": "Make", 0: "Count"})
ranking.sort_values(by=['Count'], ascending = False).reset_index(drop=True)
```
