```python
#確認entertainment頁面的範圍
```

```python
base_page = requests.get('https://www.niusnews.com')
base_page
```

```python
base_soup = BeautifulSoup(base_page.text, 'html.parser')
base_soup
```

```python
links = base_soup.find_all('a')

pages = []

for link in links:
    href = link.get('href')
    if href and '/channel/entertainment/' in href:
        pages.append(href[-3:])
start = int(min(pages))
end = int(max(pages))
```

```python
#for迴圈，順序為前向後
```

```python
def fetch_pages_for_loop(start_page, end_page):
    BASE_URL = 'https://www.niusnews.com'
    current_page = f'{BASE_URL}/channel/entertainment/{start_page}'

    for _ in range(start_page, end_page + 1):
        response = requests.get(current_page)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            for article in soup.find_all('a', class_='subject'):
                print(article.text.strip())

        else:
            print("無法訪問頁面，請檢查連結或網絡狀態。")
            break
```

```python
fetch_pages_for_loop(start, end)
```

```python
#將函式改成while迴圈，並加入若是遇到沒有值就停止，順序為後向前
```

```python
def fetch_pages_while_loop(start_page):
    
    BASE_URL = 'https://www.niusnews.com'
    
    while True:  
        current_page = f'{BASE_URL}/channel/entertainment/{start_page}'
    
        response = requests.get(current_page)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            if len(soup.find_all('a', class_='subject')) > 1:
                for article in soup.find_all('a', class_='subject'):
                    print(article.text.strip())
                start_page -= 1
            else:
                print('沒有更多資料了')
                break
        else:
            print
            break
```

```python
fetch_pages_while_loop(end)
```
